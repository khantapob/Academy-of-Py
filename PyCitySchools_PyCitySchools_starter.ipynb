```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_schools = school_data["school_name"].count()
total_schools 

#Calculate the total number of students
total_students = student_data["student_name"].count()
total_students 

#Calculate the total budget
total_budget = school_data["budget"].sum()
total_budget 

#Calculate the average math score
avg_math_score = student_data["math_score"].mean()
avg_math_score 

#Calculate the average reading score
avg_reading_score = student_data["reading_score"].mean()
avg_reading_score 

#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
avg_overall_score = (student_data["math_score"].mean() + student_data["reading_score"].mean())/2
avg_overall_score 

#Calculate the percentage of students with a passing math score (70 or greater)
pass_math = student_data.loc[(student_data["math_score"] > 69)]
pass_math = pass_math["student_name"].count()
percent_pass_math = (pass_math/total_students) *100
percent_pass_math

#Calculate the percentage of students with a passing reading score (70 or greater)
pass_reading = student_data.loc[(student_data["reading_score"] > 69)]
pass_reading = pass_reading["student_name"].count()
percent_pass_reading = (pass_reading/total_students) * 100
percent_pass_reading

#Create a dataframe to hold the above results
district_summary = pd.DataFrame({"Total Schools": [total_schools],
                                 "Total Students": [total_students],
                                 "Total Budget": [total_budget],
                                 "Average Math Score": [avg_math_score],
                                 "Average Reading Score": [avg_reading_score],
                                 "Average Overall Score": [avg_overall_score],
                                 "% Passing Math": [percent_pass_math],
                                 "% Passing Reading": [percent_pass_reading],
                                 "% Overall Passing Rate": [avg_overall_score],})



#Optional: give the displayed data cleaner formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"]= district_summary["Total Budget"].map("${:,.2f}".format)
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:

* Create a dataframe to hold the above results

```python
#School Name
#set schools as index
school_index = school_data.set_index("school_name")

#School Type
school_type = school_index["type"]
#school_type

#Total Students
total_students = school_data.groupby(["school_name"]).sum()
total_students = total_students["size"]
total_students 

#Total School Budget
total_budget = school_data.groupby(["school_name"]).sum()
total_budget = total_budget["budget"]

#Per Student Budget
per_student_budget = school_data.groupby(["school_name"]).sum()
per_student_budget = per_student_budget["budget"]/per_student_budget["size"]
#per_student_budget

#Average Math Score
#Average Reading Score
test_score = school_data_complete.groupby(["school_name"])
avg_test_score = test_score.mean()
avg_test_score

#% Passing Math
student_M_score = school_data_complete.loc[(school_data_complete["math_score"] > 69)]
student_M_score = student_M_score.groupby(["school_name"])
pass_math = (student_M_score["math_score"].count()/test_score["size"].count()) * 100
pass_math

#% Passing Reading
student_R_score = school_data_complete.loc[(school_data_complete["reading_score"] > 69)]
student_R_score = student_R_score.groupby(["school_name"])
pass_reading = (student_R_score["math_score"].count()/test_score["size"].count()) * 100
pass_reading

#Overall Passing Rate (Average of the above two)
overall_passing = (pass_math + pass_reading)/2
overall_passing


school_summary = pd.DataFrame({"School Type": school_type,
                               "Total Students": total_students,
                               "Total School Budget": total_budget,
                               "Per Student Budget": per_student_budget,
                               "Average Math Score": avg_test_score["math_score"],
                               "Average Reading Score": avg_test_score["reading_score"],
                               "% Passing Math": pass_math,
                               "% Passing Reading": pass_reading,
                               "Overall Passing Rate": overall_passing
                              })
                    

school_summary
```
